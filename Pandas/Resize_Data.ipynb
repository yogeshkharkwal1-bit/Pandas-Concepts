# MELT AND PIVOT

#### MELT - wide and long formate
1. we resize the wide data into long data
2. means increase the row count
3. decrease the column count
4. data visulization and normalization easy
5. we pass the {
6. id_var[keep in],
7. values_var[unpivot]
8. var_name[new column name]
9. value_name)[new column values]

#### Pivot - long to wide data
1. we resize the long data into wide data
2. means increase the column count
3. decrease the row count
4. we paas the...
5. index for keeping
6. columns for new values added
7. values for adding those values

```python
import pandas as pd
```

```python
df=pd.DataFrame({
    "country":["USA","USA","INDIA","INDIA"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    # "tax":["12%","13%","16%","12%"]
})
df
```

```python
mdf=df.melt( 
     id_vars=["country","year"], #inko change nhi krna he
     value_vars=["sales","profit"], #inko hum transform krna chahte he
     var_name="metrics", # sales or profit ke lie naya  naam
     value_name="values"# jo value aaegi unke lie naam
)
mdf
```

```python
# long to wide formate
mdf.pivot(
    index=["country","year"],
    columns="metrics",
    values="values"
)
```
