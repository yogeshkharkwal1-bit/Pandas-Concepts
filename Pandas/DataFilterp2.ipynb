# Query() method for filteriing


### Rules of query..

1. expression as a string ("aqi > 100")
2.  column based refrence
3. backticks fro col_name with space or special chars
4. operators(& , |,~,>=,<=,== etc.)
5. @ to reference python vars
6. chained comparison >>>("aqi > 100 & temperature >20") [["city","aqi"]]
          
    
         

```python
import pandas as pd
```

```python
df=pd.read_csv("globaldata.csv")

# Query
# df.query("aqi > 100 & temperature >20") [["city","aqi"]]
aqi_data=df.query("aqi > 100 & temperature >20") [["city","aqi"]].copy()
aqi_data
```
