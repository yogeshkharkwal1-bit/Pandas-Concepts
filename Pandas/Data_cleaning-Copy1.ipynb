```python
import pandas as pd
```

```python
df=pd.read_csv("raw_data.csv")
```

```python
df2=df.copy()
df2
```

```python
df2=df2.bfill()  #if we read the backward ,the nan is replace by thw backward value
df2
```

```python
# cleaning the data with handle dublicacy

df2.duplicated() 
# df[["country","gender"]].duplicated()
df2=df2.drop_duplicates()
df2=df2.reset_index(drop=True)
df2
```

```python
# df2.to_csv("clear_data.csv")
# Grouping and agration

# making groups with the help of perameeters

df2.groupby("country")["income"].mean()
df2.groupby("country")["income"].min()
df2.groupby("country")["income"].max()

df2.groupby("gender")["income"].mean()
df2.groupby("gender")["income"].min()
df2.groupby("gender")["income"].max()
df2.groupby("gender")["income"].sum()

# multiple comparison in one line by agg
df2.groupby("gender")["income"].agg(["mean","sum","min","max","count"])
df2.groupby("country")["income"].aggregate(["mean","sum","min","max","count"])

df2.groupby("gender")["income"].agg(Avg_Salary="mean",Sum_of_Salary="sum",min_Salary="min",max_Salary="max",avg_count="count")
df2.groupby("country")["income"].agg(Avg_Salary="mean",Sum_of_Salary="sum",min_Salary="min",max_Salary="max",avg_count="count")

df2.groupby("country").agg({
    "income":"mean",
    "age":"mean"
})

df.groupby("country").agg({
    "income":"mean",
    "age":"max"
})

df.groupby("country").agg(
    Avg_Salary=("income","mean"),
    Avg_age=("age","mean")
)


df.groupby("country").agg(
    Max_Salary=("income","max"),
    Min_age=("age","min")
)
```
