# A Data Frame is a 2D labbled array 

```python
import pandas as pd
```

```python
#data frame

# 1st method to create a data frame is dictonart

info={
    "Name":["Ram","Syam","Rohan","Sohan"],
    "Age":[21,16,19,20],
"Cgpa":[9,8.9,9.7,9.2]
 
}

df=pd.DataFrame(info)

print(df)

print(type(df)) #<class 'pandas.core.frame.DataFrame'>

print(df.index) #RangeIndex(start=0, stop=4, step=1)
print(df.columns)
```

```python
# 2nd method to create data frame is list to  list,we also pass the collum values

df=pd.DataFrame([["Aman",21],["Bob",22],["Eve",23]],columns=["Name","Age"])

print(df)
```
