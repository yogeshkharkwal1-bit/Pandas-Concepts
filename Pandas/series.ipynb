#### Index are a positions of data 

```python
import pandas as pd
```

```python
s=pd.Series([2,3,4,5,6])
print(s)
print(type(s))

print(s[0])
print(s[1])
print(s[2])
```

```python
s=pd.Series([23,45,21,45,64],index=["Bob","Charlie","Yogi","Sem","Jorden"])

print(s["Yogi"])  #named index
print(s[2])   #position index
print(s.index)
```

```python
s1=pd.Series([2,3,4,5,6],dtype="int8")
s2=pd.Series([2,3,4,5,6],dtype="int8")

print(s1)
print(s2)
print(s1+s2)
```

```python
s1=pd.Series([2,3,4,5,6],dtype="int8")

s1[0]=1
changed_s1=s1.drop(0) # nai sseries bnani pdegi agr size change krna he to

print(s1)
print(changed_s1)
```
